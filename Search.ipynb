```python
!pip install googlesearch-python
!pip install ipywidgets

from googlesearch import search
import ipywidgets as widgets
from IPython.display import display, clear_output

def search_query(btn):
    query = text.value
    num_results = int(num_results_text.value)
    
    search_results = list(search(query, num_results=num_results))
    search_results = search_results[:num_results]
    
    with result_text:
        clear_output()
        for i, result in enumerate(search_results):
            print(f"Result {i+1}: {result}")

# Create input widgets
text = widgets.Text(placeholder='Enter your search query', description='Query:')
num_results_text = widgets.IntText(value=5, description='Results:')
search_button = widgets.Button(description='Search', button_style='success')

# Define callback function for the search button
search_button.on_click(search_query)

# Customize the style of the input widgets
text.style.description_width = 'initial'
num_results_text.style.description_width = 'initial'

# Customize the style of the search button
search_button.style.button_color = 'hotpink'

# Create output widget
result_text = widgets.Output()

# Display the widgets
display(text, num_results_text, search_button, result_text)
```
